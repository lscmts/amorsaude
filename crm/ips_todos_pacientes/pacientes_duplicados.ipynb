```python
import pandas as pd
```

```python
caminho = '/home/lucasmateus/dbt-projects/projetos_lucas/ips_todos_pacientes/pacientes_ips_todos_duplicados.xlsx'
df = pd.read_excel(caminho)
df
```

```python
df['# Interno'] = df['# Interno'].str.strip()
df['Nombre'] = df['Nombre'].str.upper()
df['Apellidos'] = df['Apellidos'].str.upper()
df['#Número de identificación'] = df['#Número de identificación'].str.strip()
df.dtypes
```

```python
def custom_concat(row):
    nombre = str(row['Nombre'])
    apellidos = str(row['Apellidos'])
    fecha_nac = str(row['Fecha de nac.'])
    return f"{nombre} {apellidos} {fecha_nac}"

df['chave'] = df.apply(custom_concat, axis=1)
```

```python
df['duplicado_rut'] = df.duplicated(subset='# Interno',keep=False).astype(int)
df['duplicado_chave'] = df.duplicated(subset='chave', keep=False).astype(int)
df['duplicado_final'] = (df['duplicado_rut'] | df['duplicado_chave']).astype(int)
df['nao_nulos'] = df.apply(lambda row: row.count(), axis=1)
df
```

```python
df.to_excel('pacientes_validacao.xlsx', index=False, sheet_name='base')
```
